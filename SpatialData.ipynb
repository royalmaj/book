```python
import folium


mymap = folium.Map(location=[27.1751782, 78.0395565], width=950, height=550, zoom_start=12, tiles='OpenStreetMap')


popup_html = "<h3>Taj Mahal</h3><img src='https://upload.wikimedia.org/wikipedia/commons/thumb/1/1a/Taj_Mahal_%28Edited%29.jpeg/320px-Taj_Mahal_%28Edited%29.jpeg' width='300px'>"


folium.Marker(location=[27.1751782, 78.0395565], popup=popup_html, parse_html=True).add_to(mymap)


mymap
```
