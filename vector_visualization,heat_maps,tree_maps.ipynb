```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_pos=[0.3,0]
y_pos=[0,0]
x_direct=[1,-1]
y_direct=[1,-1]
fig,ax=plt.subplots(figsize=(7,7))
ax.quiver(x_pos,y_pos,x_direct,y_direct, color= 'b')
ax.set_title('quiver plot with two arrow')
plt.show()
```

```python
import plotly.express as px
```

```python
df=px.data.tips()
fig=px.treemap(df, path=['sex','day','time'], values='total_bill', color='day')
fig.update_layout(margin=dict(t=50,l=25,r=25,b=25))
fig.show()
```

```python

n = 100
x = np.linspace(-5, 5, n)
y = np.linspace(-5, 5, n)
X, Y = np.meshgrid(x, y)
u = np.cos(Y)
v = np.sin(X)
plt.streamplot(x, y, u, v)
plt.colorbar()
plt.show()
```
